```python
import random
import matplotlib.pyplot as plt

class VTUStudyMarkAnalysis:
    def __init__(self):
        self.name = ""
        self.usn = ""
        self.semester = ""
        self.branch = ""
        self.num_subjects = 0
        self.subjects = []

    def input_student_details(self):
        self.name = input("Enter your name: ")
        self.usn = input("Enter your USN: ")
        self.semester = input("Enter semester: ")
        self.branch = input("Enter branch: ")
        self.num_subjects = int(input("Enter the number of subjects: "))

    def input_subject_details(self):
        for i in range(self.num_subjects):
            subject_name = input(f"Enter subject {i + 1} name: ")
            credits = int(input(f"Enter credits for {subject_name}: "))
            grade_marks = int(input(f"Enter grade marks for {subject_name}: "))
            self.subjects.append({
                'name': subject_name,
                'credits': credits,
                'grade_marks': grade_marks
            })

    def calculate_total_marks(self):
        return sum(subject['grade_marks'] for subject in self.subjects)

    def calculate_percentage(self, total_marks):
        return (total_marks / (self.num_subjects * 100)) * 100

    def calculate_grade_letter(self, percentage):
        if percentage >= 90:
            return 'A+'
        elif 80 <= percentage < 90:
            return 'A'
        elif 70 <= percentage < 80:
            return 'B+'
        elif 60 <= percentage < 70:
            return 'B'
        elif 55 <= percentage < 60:
            return 'C'
        elif 50 <= percentage < 55:
            return 'P'
        else:
            return 'F'

    def calculate_grade_points(self, percentage):
        if percentage >= 90:
            return 10
        elif 80 <= percentage < 90:
            return 9
        elif 70 <= percentage < 80:
            return 8
        elif 60 <= percentage < 70:
            return 7
        elif 55 <= percentage < 60:
            return 6
        elif 50 <= percentage < 55:
            return 4
        else:
            return 0

    def analyze_weak_strong_scores(self):
        weak_scores = [subject for subject in self.subjects if subject['grade_marks'] < 60]
        strong_scores = [subject for subject in self.subjects if subject['grade_marks'] >= 60]
        return weak_scores, strong_scores

    def visualize_weak_strong_scores(self, weak_scores, strong_scores):
        labels = ['Weak Scores', 'Strong Scores']
        sizes = [len(weak_scores), len(strong_scores)]
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
        plt.title('Analysis of Weak and Strong Scores')
        plt.show()

    def generate_marksheet(self, total_marks, percentage, grade_letter, grade_points, weak_scores, strong_scores):
        print("\n----- Marksheet -----\n")
        print(f"Name: {self.name}\tUSN: {self.usn}\tSemester: {self.semester}\tBranch: {self.branch}\n")
        print("Subject\t\tCredits\tGrade Marks")
        print("------------------------------------")
        for subject in self.subjects:
            print(f"{subject['name']}\t\t{subject['credits']}\t{subject['grade_marks']}")
        print("------------------------------------")
        print(f"Total Marks: {total_marks}")
        print(f"Percentage: {percentage:.2f}%")
        print(f"Grade Letter: {grade_letter}")
        print(f"Grade Points: {grade_points}")
        print("------------------------------------")
        print("Weak Scores:")
        for subject in weak_scores:
            print(f"{subject['name']}: {subject['grade_marks']}")
        print("Strong Scores:")
        for subject in strong_scores:
            print(f"{subject['name']}: {subject['grade_marks']}")
        print("------------------------------------")

    def display_motivation(self):
        motivation_quotes = {
            'Farha Kousar': "Farha Kousar, your dedication and creativity shine through! Keep up the fantastic work on your project!",
            'John Doe': "Success is not final, failure is not fatal: It is the courage to continue that counts.",
            'Jane Smith': "Believe you can and you're halfway there.",
            # Add more quotes as needed
        }
        print("\nMotivation Quote:\n")
        print(motivation_quotes.get(self.name, "Stay motivated and keep working towards your goals!"))

    def run_analysis(self):
        self.input_student_details()
        self.input_subject_details()

        total_marks = self.calculate_total_marks()
        percentage = self.calculate_percentage(total_marks)
        grade_letter = self.calculate_grade_letter(percentage)
        grade_points = self.calculate_grade_points(percentage)

        weak_scores, strong_scores = self.analyze_weak_strong_scores()
        self.visualize_weak_strong_scores(weak_scores, strong_scores)

        self.generate_marksheet(total_marks, percentage, grade_letter, grade_points, weak_scores, strong_scores)

        self.display_motivation()

# Create an instance of the VTUStudyMarkAnalysis class and run the analysis
vtu_analysis = VTUStudyMarkAnalysis()
vtu_analysis.run_analysis()
```
